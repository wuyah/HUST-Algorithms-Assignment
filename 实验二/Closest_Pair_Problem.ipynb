```python
# 在 0 - 5000上生成size大小的点对
def randPoint(size:int):
    c = [[0,0] for _ in range(size)]
    for i in range(size):
        c[i][0],c[i][1] = rd.randint(0,10000), rd.randint(0,10000)
    return c
```

```python
points = randPoint(100)
points_np = np.array(points)
```

```python
def distance(a,b):
    return ((a[0]-b[0])**2 + (a[1]-b[1])**2)**(0.5)
```

```python
distance(points[0],points[1])
```

```python
def dummy(points):
    n = len(points)
    mis = distance(points[0],points[1])
    for i in range(n):
        for j in range(i+1, n):
            mis = min(distance(points[i],points[j]),mis)
    return mis
```

```python
def fix_py(px, py):
    temp = []
    l, r = px[0][0],px[-1][0]
    for x, y in py:
        if l<=x<=r:
            temp.append([x,y])
    return np.array(temp)
        
```

```python
def find_min_mid(radius, center, py):
    l, r = center - radius, center + radius
    res = []
    for x, y in py:
        if l<=x <=r:
            res.append([x,y])
    n = len(res)
    if n <= 1:
        return float('inf')
    try:
        mis = distance(res[0],res[1])
    except:
        print(n)
        print("Error")

    for i in range(n):
        if i+8>n:
            for j in range(i, n):
                for k in range(j+1, n):
                    mis = min(distance(res[j],res[k]),mis)
        else:
            for j in range(i+1, i+8):
                mis = min(distance(res[i],res[j]),mis)
    return mis
```

```python
def closest_pair_problem(points):
    px = points[points[:,0].argsort()]
    py = points[points[:,1].argsort()]
    def merge(px, py):
        n = len(px)
        if n <= 3:
            mis = distance(px[0],px[1])
            pos = [px[0][0],px[0][1],px[1][0],px[1][1]]
            for i in range(n):
                for j in range(i):
                    if distance(px[i],px[j]) < mis:
                        pos[0], pos[1] = px[i][0],px[i][1]
                        pos[2], pos[3] = px[j][0],px[j][1]
            return mis
        mid = n//2
        d1 = merge(px[:mid+1], fix_py(px[:mid],py))
        d2 = merge(px[mid:], fix_py(px[:mid],py))
        d = min(d1, d2)
        ms = min(d, find_min_mid(d, px[mid][0],py))
        temp_y = np.linspace(0,10000)
        temp_x_mid = [px[mid][0]] * len(temp_y)
        plt.plot(temp_x_mid,temp_y)
        plt.scatter(px[:,0], px[:,1])
        plt.xlim(0,10000)
        plt.ylim(0,10000)
        plt.show()
        return ms
    return merge(px,py)
```
