```python
# import numpy as np
import random as rd
from typing import List
```

```python
def random_16(down,up):
    m = [[0]*16 for _ in range(16)]
    for i in range(16):
        for j in range(16):
            m[i][j] = rd.randint(down,up)
    return m
```

```python
# 辅助函数, 将矩阵划分为四个更小的矩阵并且返回
def divide(a:List[List[int]]):
    n=len(a)//2
    a11,a12,a21,a22=[[0]*n for _ in range(n)],[[0]*n for _ in range(n)],[[0]*n for _ in range(n)],[[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            a11[i][j], a12[i][j], a21[i][j], a22[i][j]=a[i][j], a[i][j+n], a[i+n][j], a[i+n][j+n]
    return (a11,a12,a21,a22)
```

```python
def mt_add(a,b,f):
    n = len(a)
    c = [[0]*n for _ in range(n)]
    if f:
        for i in range(n):
            for j in range(n):
                c[i][j] = a[i][j]+b[i][j]
    else:
        for i in range(n):
            for j in range(n):
                c[i][j] = a[i][j]-b[i][j]
    return c
```

```python
def mt_combine(a,b,c,d):
    
    m = len(a)
    n = m * 2
    e = [[0]*n for _ in range(n)]
    for i in range(m):
        for j in range(m):
            e[i][j] = a[i][j]
            e[i][j+m] = b[i][j]
            e[i+m][j] = c[i][j]
            e[i+m][j+m] = d[i][j]
    return e
```

```python
def Strssen(a,b):
    n = len(a)
    c = [[0]*n for _ in range(n)]
    if n == 1:
        c[0][0] = a[0][0]*b[0][0]
        return c
    (a11,a12,a21,a22)=divide(a)
    (b11,b12,b21,b22)=divide(b)
    (c11,c12,c21,c22)=divide(c)
    s1=mt_add(b12,b22,0)
    s2=mt_add(a11,a12,1)
    s3=mt_add(a21,a22,1)
    s4=mt_add(b21,b11,0)
    s5=mt_add(a11,a22,1)
    s6=mt_add(b11,b22,1)
    s7=mt_add(a12,a22,0)
    s8=mt_add(b21,b22,1)
    s9=mt_add(a11,a21,0)
    s10=mt_add(b11,b12,1)
    p1=Strssen(a11,s1)
    p2=Strssen(s2,b22)
    p3=Strssen(s3,b11)
    p4=Strssen(a22,s4)
    p5=Strssen(s5,s6)
    p6=Strssen(s7,s8)
    p7=Strssen(s9,s10)
    c11=mt_add(mt_add(mt_add(p5,p4,1),p2,0),p6,1)
    c12=mt_add(p1,p2,1)
    c21=mt_add(p3,p4,1)
    c22=mt_add(mt_add(mt_add(p5,p1,1),p3,0),p7,0)
    c = mt_combine(c11,c12,c21,c22)
    return c
```
